# How to Analyze DataFrames ?

# .head()
digunakan untuk menampilkan sejumlah baris pertama dari sebuah DataFrame. Contoh data.head(10)

```python
data.head()
```

# .tail()

digunakan untuk menampilkan sejumlah baris terakhir dari sebuah DataFrame. Contoh data.tail(10)

```python
data.tail(3)
```

# .shape

Menampilkan total row dan columns dari dataframe

```python
data.shape
```

# .index
digunakan untuk mengakses indeks dari sebuah DataFrame atau Series. Indeks ini adalah label baris yang memungkinkan Anda untuk mengidentifikasi dan mengakses data dalam DataFrame atau Series.

```python
data.index
```

# .columns
Digunakan untuk mengakses label kolom (nama kolom) dari sebuah DataFrame.

```python
data.columns
```

# .dtypes
Digunakan untuk menampilkan tipe data dari setiap kolom dalam sebuah DataFrame 

```python
data.dtypes
```

# .unique()
Menampilkan nilai-nilai unik (tanpa duplikat) dan hanya bisa menampilkan satu kolom saja

```python
data['Weather'].unique()
```

# .nunique()
Menampilkan total data unik(Tanpa duplikat) dan hanya bisa menampilkan satu kolom saja

```python
data['Weather'].nunique()
```

# .count 
Digunakan untuk menghitung jumlah nilai non-NaN (bukan hilang/kosong) dalam sebuah Series atau DataFrame.

```python
data.count()
```

# .value_counts
Digunakan untuk menghitung frekuensi kemunculan setiap nilai unik dalam sebuah kolom

```python
data['Weather'].value_counts()
```

# .info()
Digunakan untuk mendapatkan ringkasan informasi penting tentang sebuah DataFrame

```python
data.info()
```
