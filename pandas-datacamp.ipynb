## Menghitung menggunakan.value_counts()
Sering kali Anda akan bekerja dengan nilai kategoris, dan Anda ingin menghitung jumlah observasi yang dimiliki setiap kategori dalam satu kolom. Nilai kategori dapat dihitung menggunakan metode <b>value_counts()</b>

```python
df.head(3)
```

```python
df['Outcome'].value_counts()
```

Dengan menambahkan argumen <b>value_counts(normalize = True)</b>, Anda akan mendapatkan proporsi, bukan jumlah mutlak.

```python
df['Outcome'].value_counts(normalize = True)
```

Anda dapat mematikan pengurutan otomatis hasil dengan menggunakan argumen <b>sort (nilai bawaannya adalah True)</b>. Pengurutan standar dilakukan berdasarkan jumlah, dari yang terbesar ke yang terkecil.

```python
df['Outcome'].value_counts(sort=False)
```

Fungsi <b>.value_counts()</b> bisa diterapkan pada objek DataFrame dan kolom-kolom tertentu, bukan hanya pada satu kolom saja. Misalnya, di sini kita menerapkan .value_counts() pada df dengan argumen <b>subset</b>, yang menggunakan daftar kolom sebagai input.

```python
df.value_counts(subset=['Pregnancies', 'Outcome'])
```
