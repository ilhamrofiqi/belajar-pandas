## Memeriksa nilai yang hilang di pandas dengan .isnull()

## Memfilter, Memotong dan Mengekstrak Data di pandas
Paket pandas menawarkan beberapa cara untuk mengurutkan, membuat subset, memfilter, dan mengisolasi data dalam DataFrames Anda.

## Memfilter data
Untuk mengurutkan DataFrame berdasarkan kolom tertentu:

```python
df.sort_values(by="Age", ascending=False, inplace=True)
```

```python
df.head()
```

```python
df.sort_values(by=["Age", "Glucose"], ascending=[False, True], inplace=True)
```

```python
df.head()
```

## Mengatur ulang indeks
Jika Anda memfilter atau mengurutkan DataFrame, indeks Anda mungkin tidak selaras, Gunakan .reset_index() untuk memperbaikinya:

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

## Memfilter data menggunakan kondisi
Untuk mengekstrak data berdasarkan suatu kondisi:

```python
df[df["BloodPressure"] > 100].sort_values(by="BloodPressure", ascending=False).head(5)
```

## Mengisolasi satu kolom menggunakan[ ] 
Anda dapat mengisolasi satu kolom menggunakan tanda kurung siku [ ]dengan nama kolom di dalamnya.

```python
df["Age"]
```

## Mengisolasi dua atau lebih kolom menggunakan [[]]
Anda juga dapat memberikan daftar nama kolom di dalam tanda kurung siku untuk mengambil lebih dari satu kolom.

```python
df[['Pregnancies','Age']]
```

## Mengisolasi satu baris menggunakan []
Satu baris dapat diambil dengan memasukkan rangkaian boolean dengan satu True nilai

```python
df.head()
```

```python
df[df.index==2]
```

## Mengisolasi dua atau lebih baris menggunakan [ ] 
Demikian pula, dua baris atau lebih dapat dikembalikan menggunakan .isin() metode, bukan == operator.

```python
df[df.index.isin(range(2,5))]
```
